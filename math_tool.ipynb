```python
import math
import numpy as np
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
def dtanh(x):
    tmp = np.tanh(x)
    return 1 - tmp * tmp
```

```python
def sigmoid(x):
    return 1 / (1 + np.exp(-1 * x))
```

```python
def dsigmoid(x):
    y = sigmoid(x)
    return y * (1 - y)
```

```python
def d_tanh_sigmoid(x):
    y = sigmoid(x)
    d = dtanh(y) * y * (1 - y)
    return d
```

```python
def dtanh_sigmoid(x):
    y = sigmoid(x)
    return dtanh(y)
```

```python
x = np.linspace(-10, 10, 100)
```

```python
plt.plot(x, d_tanh_sigmoid(x), label = 'd_tanh_sigmoid')
plt.plot(x, dtanh_sigmoid(x), label = 'dtanh_sigmoid')
plt.plot(x, sigmoid(x), label = 'simoid')
plt.plot(x, np.tanh(x), label = 'tanh')
plt.legend()
plt.show()
```

```python
def elipse_func(x, y):
    return np.power(x, 2) + np.power(y, 2)
```

```python
def dtanh_sigmoid_xy(x, y):
    return dtanh(x) * sigmoid(y)
```

```python
#z = elipse_func(x, y)
z = dtanh_sigmoid_xy(x, y)
zz = np.log10(z)
d = dict()
for i in zz:
    for j in i:
        j = int(j)
        if j not in d:
            d[j] = 1
        else:
            d[j] += 1
dd = sorted(d.items(), key = lambda x:x[1], reverse = True)
for k, v in dd:
    print(str(k) + ':' +str(v))
```

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D

fig=plt.figure()
ax=Axes3D(fig)

X=np.arange(-4,4,0.25)
Y=np.arange(-4,4,0.25)
X,Y=np.meshgrid(X,Y)
R=np.sqrt(X ** 2 + Y ** 2)
Z=np.sin(R)

ax.plot_surface(X,Y,Z,rstride=1,cstride=1,cmap=plt.get_cmap('rainbow'))

ax.contourf(X,Y,Z,zdir='z',offset=-2,cmap=plt.get_cmap('rainbow'))

plt.show()
```
